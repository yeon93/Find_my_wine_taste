## 분류모델 생성
+ n_clusters=100

```python
import pickle
import pandas as pd
```

```python
data = pd.read_csv('wine_cluster_100.csv', index_col=0)

data.head(3)
```

```python
data.fillna('NaN', inplace=True)
print(data.shape)

samples = data.sample(3, random_state=2).values
samples
```

```python
with open('model_100clusters.pkl', 'rb') as file: 
    model = pickle.load(file)
    
for i in samples:
    X_test = i[1:].reshape(-1,1)
    y_pred = model.predict(X_test)[0]
    print(f'실제 클러스터 : {i[0]}, 예상 클러스터 : {y_pred}')
```

```python
y_preds = []
for i in range(len(data)):
    X_test = data.iloc[i][1:].values.reshape(-1,1)    
    y_pred = model.predict(X_test)[0]
    y_preds.append(y_pred)
```

```python
counts = []
for i in range(100):
    counts.append(y_preds.count(i))

len(data), counts
```
